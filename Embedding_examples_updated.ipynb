**BAG OF WORDS**

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Sample text
documents = [
    "I love programming",
    "Python is great for data science",
    "I love Python"
]
```

```python
# Initialize CountVectorizer
vectorizer = CountVectorizer()
```

```python
# Fit and transform the documents
X = vectorizer.fit_transform(documents)
```

```python
# Display the BoW representation
print(vectorizer.get_feature_names_out())
print(X.toarray())
```

**TF_IDF Vectorizer**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Sample text
documents = [
    "I love python programming",
    "Python is great for data science",
    "I love Python",
    "Python is one of my favorite language"
]
```

```python
# Fit and transform the documents
X = vectorizer.fit_transform(documents)
```

```python
# Display the TF-IDF representation
print(vectorizer.get_feature_names_out())
print(X.toarray())
```

**Why TF-IDF is Useful:
Frequency (TF) captures the importance of the word within a document.
Inverse Frequency (IDF) reduces the weight of words that are common across many documents, so it emphasizes terms that are more unique and potentially more meaningful.**

```python
import pandas as pd
import numpy as np
```

```python
corpus = [
    "I love python programming",
    "Python is great for data science",
    "I love Python",
    "Python is one of my favorite language"
]

```

```python
words_set = set()

for doc in  corpus:
    words = doc.split(' ')
    words_set = words_set.union(set(words))
print(words)
print('Number of words in the corpus:',len(words_set))
print('The words in the corpus: \n', words_set)
```

```python
n_docs = len(corpus)         #¬∑Number of documents in the corpus
n_words_set = len(words_set) #¬∑Number of unique words in the

print("the no of documentys is ", n_docs)
print(n_words_set)
```

```python
df_tf = pd.DataFrame(np.zeros((n_docs, n_words_set)), columns=['I', 'Python', 'of', 'science', 'my', 'data', 'great', 'is', 'one', 'love', 'programming', 'python', 'language', 'favorite', 'for'])
df_tf
```

```python

#Compute Term Frequency (TF)
for i in range(n_docs):
    words = corpus[i].split(' ') # Words in the document
    for w in words:
        df_tf[w][i] = df_tf[w][i] + (1 / len(words))

df_tf
```

```python
print("IDF of: ")

idf = {}

for w in words_set:
    k = 1    # number of documents in the corpus that contain this word

    for i in range(n_docs):
        if w in corpus[i].split():
            k += 1

    idf[w] =  np.log10(n_docs / k)

    print(f'{w:>15}: {idf[w]:>10}' )
```

```python
df_tf_idf = df_tf.copy()

for w in words_set:
    for i in range(n_docs):
        df_tf_idf[w][i] = df_tf[w][i] * idf[w]

df_tf_idf
```
